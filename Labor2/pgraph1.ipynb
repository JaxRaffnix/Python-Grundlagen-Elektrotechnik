# Measure $S$

$S = \frac{R_z || R_l}{R_z || R_l + Rv}$

$U_E = 12 V$

```python
outvoltage_involtage_S = pd.read_excel(r"measurement.xlsx", sheet_name="3-3-b")

# Ue_delta = outvoltage_involtage_S["Ue in V"].max() - outvoltage_involtage_S["Ue in V"].min()
Ua_delta = outvoltage_involtage_S["Ua in V"].max() - outvoltage_involtage_S["Ua in V"].min()
Ia_delta = outvoltage_involtage_S["Ia in mA"].max() - outvoltage_involtage_S["Ia in mA"].min()

Rz = abs(Ua_delta / Ia_delta)

Rzl = (Rz * Rl)/(Rz + Rl)
S_measured = Rzl / (Rzl + Rv)
print("S gemessen = ", S_measured)
```
