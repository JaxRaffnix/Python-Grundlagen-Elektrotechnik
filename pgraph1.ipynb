```python
import pandas as pd
import matplotlib.pyplot as plt

diode_curve = pd.read_excel(r"measurement.xlsx", sheet_name="3-2-a")

diode_curve_x = list(diode_curve["Uz in V"])
diode_curve_y = list(diode_curve["Iz in mA"])

plt.plot(diode_curve_x, diode_curve_y,)
plt.show()
```

```python
diode_tang_import = pd.read_excel(r"measurement.xlsx", sheet_name="3-2-d")

diode_tang_Iz = list(diode_tang_import["Iz in mA"])
diode_tang_Uz = list(diode_tang_import["Uz in V"])

Uz_delta = max(diode_tang_Uz) - min(diode_tang_Uz)
Iz_delta = max(diode_tang_Iz) - min(diode_tang_Iz)

Iz0 = -80
Uz0 = -4.781
tang_slope = Iz_delta / Uz_delta
increment = 0.11
tang_next = tang_slope*increment + Iz0

y = [-80, tang_next]
x = [Uz0, Uz0 + increment]

plt.plot(x,y)
plt.plot(diode_curve_x, diode_curve_y,)
plt.show()
```

```python
Rz = abs(Uz_delta / Iz_delta)
print("Rz = ",Rz,"Ohm")
```

```python
Rv = 50
Rl = 220

Rzl = (Rz * Rl)/(Rz + Rl)
S = Rzl / (Rzl + Rz)
print("S = ",S)
```
