```python
task_3_2_d_import = pd.read_excel(r"3-2-d.xlsx", sheet_name="Tabelle1")

task_3_2_d_x = list(task_3_2_d_import["Iz in mA"])
task_3_2_d_y = list(task_3_2_d_import["Uz in V"])

Uz_delta = max(task_3_2_d_y) - min(task_3_2_d_y)
Iz_delta = max(task_3_2_d_x) - min(task_3_2_d_x)

Rz = abs(Uz_delta / Iz_delta)

print(Rz)
```

```python
Rv = 50
Rl = 200

Rzl = (Rz * Rl)/(Rz + Rl)
S = Rzl / (Rzl + Rz)
print(S)
```
